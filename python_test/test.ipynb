```python
import numpy as np
import scipy.sparse as spa
np.random.seed(1)

n = 10
m = int(n * 10)

problem = {}
# Generate problem data
problem['n'] = int(n)
problem['m'] = m
Q = spa.random(n, n, density=0.15,
               data_rvs=np.random.randn,
                 format='csr')
problem['Q'] = Q.dot(Q.T).tocsr() + 1e-02 * spa.eye(n)
problem['c'] = np.random.randn(n)
problem['A'] = spa.random(m, n, density=0.15,
                    data_rvs=np.random.randn,
                    format='csr')
v = np.random.randn(n)   # Fictitious solution
delta = np.random.rand(m)  # To get inequality
problem['u'] = problem['A']@v + delta
problem['l'] = - np.inf * np.ones(m)  # self.u - np.random.rand(m)
```

```python
Q = problem['Q']
Q_value = Q.data.tolist()
Q_rowStart = Q.indptr.tolist()
Q_column = Q.indices.tolist()

A = problem['A']
A_value = A.data.tolist()
A_rowStart = A.indptr.tolist()
A_column = A.indices.tolist()

```
